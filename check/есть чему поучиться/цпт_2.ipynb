```python
import matplotlib.pyplot as plt
import numpy as np
#Задание констант и переменных для объема n=5
mx=0.5
dx=1.0/12.0
n=5
s=0
r=[]
mu=mx
#стандартное отклонение
sigma=np.sqrt(1/(12.0*n))
for i in range(1000):
    for x in range(n):
        s=s+np.random.random()
    r.append(s/n)
    s=0
count, bins, ignored=plt.hist(r,bins=50,normed='true')
#задание названия графика и подписей к осям
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Histogramma raspred n=5')

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()

#Задание констант и переменных для объема n=10
n=10
s=0
r=[]
mu=mx
#стандартное отклонение
sigma=np.sqrt(1/(12.0*n))
for i in range(1000):
    for x in range(n):
        s=s+np.random.random()
    r.append(s/n)
    s=0
count, bins, ignored=plt.hist(r,bins=50,normed='true')
#задание названия графика и подписей к осям
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Histogramma raspred n=10')
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()

#Задание констант и переменных для объема n=50
n=50
s=0
r=[]
mu=mx
#стандартное отклонение
sigma=np.sqrt(1/(12.0*n))
for i in range(1000):
    for x in range(n):
        s=s+np.random.random()
    r.append(s/n)
    s=0
count, bins, ignored=plt.hist(r,bins=50,normed='true')
#задание названия графика и подписей к осям
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Histogramma raspred n=50')

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()
#Выводы: с ростом объема n точность аппроксимации повышается
```
