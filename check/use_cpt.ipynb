```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts
%matplotlib inline
```

```python
dweibull_rv=sts.dweibull(3,3,1)#выбрала распределение
```

```python
sample=dweibull_rv.rvs(1000)#сгенерировала выборку объёма 1000
```

```python
plt.hist(sample,)#построила гистограмму выборки
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

```python
plt.hist(sample, normed='True', label='hist')#нарисовала поверх неё теоретическую плотность распределения случайной величины 

x=np.linspace(0,6,100)
pdf=dweibull_rv.pdf(x)
plt.plot(x, pdf, label='th_PDF',alpha=0.5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

```python
new=np.ones(1000)#сгенерировала 1000 выборок объёма 5 и построила гистограмму распределения их выборочных средних. 
n=5
for i in range(1000):
    sample=dweibull_rv.rvs(size=n)
    summa_vyborki=np.sum(sample)
    srednee_vyborki=summa_vyborki/n
    new[i]=new[i]*srednee_vyborki

plt.hist(new, normed='True',label='hist')
plt.ylabel('number of samples')
plt.xlabel('$x$')
```
