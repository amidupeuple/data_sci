```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

```python
#Распределение Фишера
d1 = 4; d2 = 7 #степени свободы
distr = sts.f(d1, d2)
mu = d2/(d2-2) #матожидание
disp = ((2.0 * (d2**2) * (d1+d2-2.0))/(d1*(d2-2.0)**2 * (d2-4.0))) #дисперсия
```

```python
#генерируем выборку
sample = distr.rvs(size=1000)
plt.hist(sample, normed=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
pdf = distr.pdf(x)
plt.plot(x, pdf)
```

```python
#получаем массив выборочных средних
def get_average(n):
    average = []
    for i in range(0, 1000):
        sample = distr.rvs(size=n)
        average.append(sum(sample)/n)
    return average

#рисуем
def fisher(n):
    average = get_average(n)    
    norm_rv = sts.norm(loc=mu, scale=math.sqrt(disp/n))
    plt.hist(average, normed=True)    
    x = np.linspace(0,10,50)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

```python
fisher(5)
```
