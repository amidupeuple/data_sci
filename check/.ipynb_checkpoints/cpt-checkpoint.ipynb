 ## Центральная предельная теорема своими руками

```python
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
import math
%matplotlib inline
```

Сгенерируем выборку объемом 1000 из Экспоненциального распределения с $\lambda=1.0$

```python
e_lambda=1.0
rv = sts.expon(loc=e_lambda)
sample = rv.rvs(1000)
```

И отобразим ее гистограмму, а также теоретическую плотность распределения

```python
plt.hist(sample, bins=10, normed=True, label="hist")
x = np.linspace(-4,4,100)
pdf = rv.pdf(x)
plt.plot(x, pdf, label='theoretical PDF',alpha=1)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
```

Сгенерируем 1000 выборок с числом элементов $n=5$ и по каждой выборке вычислим выборочное среднее

```python
n=5
s = [(sum(rv.rvs(n))/float(n)) for x in xrange(0,10000)]
```

Приближаем выборочные средние при помощи нормального распределения $N(\mu,\sigma^2/n)$ , где для экспоненциального распределения $\mu=\lambda^{-1}$ а $\sigma=\lambda^{-2}$

```python
plt.hist(s, bins=10, normed=True, label='VS hist')
x = np.linspace(-4,4,100)
norm_rv = sts.norm(loc=e_lambda**-1, scale=math.sqrt((e_lambda**-4)/n))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical PDF',alpha=1)
plt.xlabel('$x$')
plt.ylabel('$T(x)$')
plt.legend(loc='upper left')
```

Повторим все то же самое для выборок с $n=20$

```python
n=20
s = [(sum(rv.rvs(n))/float(n)) for x in xrange(0,1000)]
```

```python
plt.hist(s, bins=10, normed=True, label='VS hist')
norm_rv = sts.norm(loc=e_lambda**-1, scale=math.sqrt(e_lambda**-4/n))
x = np.linspace(-4,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical PDF',alpha=1)
plt.xlabel('$x$')
plt.ylabel('$T(x)$')
plt.legend(loc='upper left')
```

Повторим все то же самое для выборок с $n=40$

```python
n=80
s = [sum(rv.rvs(n))/float(n) for x in xrange(0,1000)]
```

```python
plt.hist(s, bins=10, normed=True, label='VS hist')
norm_rv = sts.norm(loc=e_lambda**-1, scale=math.sqrt(e_lambda**-4/n))
x = np.linspace(-4,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical PDF',alpha=1)
plt.xlabel('$x$')
plt.ylabel('$T(x)$')
plt.legend(loc='upper left')
```

Как мы убедились, Центральная Предельная Теорема выполняется для заданной выборки, апроксимируя средние значения тем точнее, чем длинее выборки из распределения.
