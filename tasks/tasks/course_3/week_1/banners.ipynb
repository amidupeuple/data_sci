```python
import pandas as pd
```

```python
fsq_df = pd.read_csv('checkins.csv')
```

```python
fsq_df.head()
```

```python
fsq_df = fsq_df[pd.notnull(fsq_df['latitude'])]
```

```python
print fsq_df.shape
```

```python
from sklearn.cluster import MeanShiftnShift
```

```python
train_df = fsq_df[0:100000]
print train_df.shape
```

```python
train_df.drop(['id', 'user_id', 'venue_id', 'create', 'd_at'], axis=1, inplace=True)
print train_df.shape
```

```python
clustering = MeanShift(bandwidth=0.1)
y = clustering.fit_predict(train_df.as_matrix())
```

```python
y[0:5]
```

```python
def fillDict(d, cl_y):
    if cl_y in d:
        d[cl_y] += 1
    else:
        d[cl_y] = 1   
```

```python
dic = {}
map(lambda cl_y: fillDict(dic, cl_y), y)
```

```python
list_of_suitable_cluters = []

for k, v in dic.iteritems():
    if v > 15:
        list_of_suitable_cluters.append(k)
        
print len(list_of_suitable_cluters)        
```

```python
train_df.head()
```

```python
lats = map(lambda x: x, train_df['latitude'])
longs = map(lambda x: x, train_df['longitude'])
```

```python
list_of_centers = []
for cluster_y in list_of_suitable_cluters:
    indices = [i for i, x in enumerate(y) if x == cluster_y]
    
    tmp_lat = 0.
    tmp_long = 0.
    for idx in indices:
        tmp_lat += lats[idx]
        tmp_long += longs[idx]
        
    list_of_centers.append((tmp_lat/len(indices), tmp_long/len(indices)))
```

```python
len(list_of_centers)
```

```python
list_of_centers[:3]
```

```python
list_offices_coordinates = [(33.751277, -118.188740),
                            (25.867736, -80.324116),
                            (51.503016, -0.075479),
                            (52.378894, 4.885084),
                            (39.366487, 117.036146),
                            (-33.868457, 151.205134)]
```

```python
import numpy as np
```

```python
list_distances = []
for cluster in list_of_centers:
    for office in list_offices_coordinates:
        dist = np.linalg.norm(np.array(office) - np.array(cluster))
        list_distances.append((cluster, dist))
print len(list_distances)
    
```
